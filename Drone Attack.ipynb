```python
#import libraries
import geopandas as gpd
import pandas as pd
import folium
```

```python
#create a map on Pakistan location
pakistan_map = folium.Map(location = [30.3753, 69.3451],zoom_start=6)
```

```python
#import Google Drive and mount the entire drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#load the dataset (csv file) into the Pandas dataframe
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DroneAttack.csv')
```

```python
#calculate total number of cases using a loop
t = 0
for c in df['deaths']:
  t = int(c)+int(t)
```

```python
for lat,long,dat,city,deaths in zip(list(df['latitude']),list(df['longitude']),list(df['date']),list(df['City']),list(df['deaths'])):
    folium.CircleMarker(location = [lat,long],
                       radius = (deaths/t)*100,
                       color = 'red',
                       fill = False,
                       fill_color = 'blue').add_to(pakistan_map)
    xyz = '<strong>City:  ' + city + '</strong><br>' + 'Total deaths: ' + str(deaths) + '<br>'
    iframe = folium.IFrame(xyz, width=200, height=150)
    popupData = folium.Popup(iframe, max_width=500)
    folium.Marker(location = [lat,long], popup=popupData).add_to(pakistan_map)

pakistan_map
```

```python
#plot co-ordinates values
dronesmap = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(x=df.longitude, y=df.latitude))
dronesmap.plot()
```

```python
#plot co-ordinates values on the map
displaymap = dronesmap.set_crs(epsg=4326)
displaymap.crs
displaymap.explore()
```
