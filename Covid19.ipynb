```
#import libraries
import folium
import pandas as pd
```

```
#create a map on Pakistan location
pakistan_map = folium.Map(location = [31.1704, 72.7097],zoom_start=8,tiles='OpenStreetMap')
```

```
#import Google Drive and mount the entire drive
from google.colab import drive
drive.mount('/content/drive')
```

```
#load the dataset (csv file) into the Pandas dataframe
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/covid.csv')
df
```

```
#calculate total number of cases using a loop
t = 0
for c in df['Total Cases']:
  t = int(c)+int(t)
```

```
#mark circles on the map and display the map
for state,TotalCases,Latitude,Longitude in zip(list(df['States']),list(df['Total Cases']),list(df['Latitude']),list(df['Longitude'])):
    folium.CircleMarker(location = [Latitude,Longitude],
                       radius = (TotalCases/t)*100,
                       color = 'red',
                       fill = False,
                       fill_color = 'blue').add_to(pakistan_map)
    xyz = '<strong>State:  ' + state + '</strong><br>' + 'Total Cases: ' + str(TotalCases) + '<br>'
    iframe = folium.IFrame(xyz, width=200, height=150)
    popupData = folium.Popup(iframe, max_width=500)
    folium.Marker(location = [Latitude,Longitude], popup=popupData).add_to(pakistan_map)

pakistan_map
```
